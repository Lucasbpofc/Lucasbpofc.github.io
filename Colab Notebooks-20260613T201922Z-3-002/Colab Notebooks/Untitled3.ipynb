```python
import pandas as pd
import numpy as np

# 1. Definir uma semente aleatória para que o sorteio seja sempre o mesmo
np.random.seed(42)

# 2. Criar o DataFrame com os dados do exercício do livro
dados = pd.DataFrame({
    'Obs': [1, 2, 3, 4, 5, 6],
    'X1': [1, 1, 0, 5, 6, 4],
    'X2': [4, 3, 4, 1, 2, 0]
})

# 3. Fazer o sorteio aleatório dos grupos (1 ou 2) para as 6 observações
# O size=6 define o número de sorteios e o p=[0.5, 0.5] garante chances iguais para os grupos
dados['Grupo_Inicial'] = np.random.choice([1, 2], size=6, replace=True, p=[0.5, 0.5])

# 4. Visualizar o banco de dados com os grupos sorteados
print("--- Tabela com os Grupos Sorteados ---")
print(dados.to_string(index=False))

# 5. Calcular os Centroides Iniciais (a média de X1 e X2 para cada grupo)
print("\n--- Centroides Iniciais (Médias por Grupo) ---")
centroides = dados.groupby('Grupo_Inicial')[['X1', 'X2']].mean().reset_index()
print(centroides.to_string(index=False))
```
