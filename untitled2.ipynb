```python
# 1. Carrega as fronteiras oficiais e isola a Zona da Mata
print("Buscando limites espaciais da Zona da Mata...")
meso_mg = geobr.read_meso_region(code_meso="MG", year=2020)
zona_da_mata = meso_mg[meso_mg['name_meso'] == 'Zona da Mata']

# Extrai a "caixa" de coordenadas extremas (min_lon, min_lat, max_lon, max_lat)
min_lon, min_lat, max_lon, max_lat = zona_da_mata.total_bounds

# 2. Carrega o cubo de dados e já recorta na mosca
caminho_input = "/content/drive/MyDrive/DOUTORADO/DOUTORADO/br_maximas_anuais_oni_1961_2025.nc"
print(f"Lendo e recortando o NetCDF na região alvo...")

ds_br = xr.open_dataset(caminho_input)
ds_ordenado = ds_br.sortby(['latitude', 'longitude'])

# Recorte espacial inteligente usando os limites da Zona da Mata (com uma folguinha de 0.2 graus)
ds_zm = ds_ordenado.sel(
    longitude=slice(min_lon - 0.2, max_lon + 0.2),
    latitude=slice(min_lat - 0.2, max_lat + 0.2)
)

print(f"✅ Malha da Zona da Mata extraída: {len(ds_zm.latitude)} latitudes x {len(ds_zm.longitude)} longitudes.")
```

```python
import xarray as xr

# 1. Monta os caminhos dos três arquivos em ordem cronológica
caminho_base = "/content/drive/MyDrive/DOUTORADO/"
arquivos_diarios = [
    caminho_base + "pr_19610101_19801231_BR-DWGD_UFES_UTEXAS_v_3.2.4.nc",
    caminho_base + "pr_19810101_20001231_BR-DWGD_UFES_UTEXAS_v_3.2.4.nc",
    caminho_base + "pr_20010101_20251231_BR-DWGD_UFES_UTEXAS_v_3.2.4.nc"
]

print("Abrindo e costurando os três arquivos diários do Xavier...")
# O xarray junta os três arquivos em um único dataset "preguiçoso" (lazy load)
ds_diario_br = xr.open_mfdataset(arquivos_diarios, combine='by_coords')

print("Recortando para a região da Zona da Mata...")
# 2. Faz o recorte espacial usando as coordenadas que você já extraiu com o geobr
ds_diario_zm = ds_diario_br.sel(
    longitude=slice(min_lon - 0.2, max_lon + 0.2),
    latitude=slice(min_lat - 0.2, max_lat + 0.2)
)

# 3. Carrega apenas os dados da Zona da Mata na memória RAM para os cálculos ficarem muito mais rápidos
ds_diario_zm = ds_diario_zm.load()

print(f"✅ Base diária montada e pronta na memória!")
print(f"Total de dias ({ds_diario_zm.time.values[0].astype('datetime64[D]')} até {ds_diario_zm.time.values[-1].astype('datetime64[D]')}): {len(ds_diario_zm.time)}")
```

```python
# Filtra dias secos (chuva < 1mm) para o cálculo do limiar não ficar enviesado
dias_com_chuva = ds_diario_zm['pr'].where(ds_diario_zm['pr'] > 1.0)

# Calcula o percentil 95 (limiar POT)
limiar_pot = dias_com_chuva.quantile(0.95).values
print(f"Limiar POT (95%) para a Zona da Mata: {limiar_pot:.2f} mm/dia")
```
